```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
```

```python
length = 100. #sec
freq = 15. #Hz
x1 = np.linspace(-np.pi*10, np.pi*10, length*freq)
wave = np.sin(x1)
t1 = np.linspace(0,length/freq,length*freq)
len(wave), len(t1)
```

```python
length = 100. #sec
freq = 20. #Hz
x2 = np.linspace(-np.pi*10, np.pi*10, length*freq)
waveNew = np.sin(x2)
waveNew2 = np.copy(waveNew)
N = 4
for i in np.arange(1, len(waveNew)):
    if i%N == 0:
        waveNew2[i] = np.nan
t2 = np.linspace(0,length/freq,length*freq)
len(waveNew), len(waveNew2), len(t2)
```

```python
length = 100. #sec
freq = 20. #Hz
x2 = np.linspace(-np.pi*10, np.pi*10, length*freq)
waveNew = np.sin(x2)
waveNew3 = np.copy(waveNew)
dropList = np.array([])
N = 4
for i in np.arange(0, len(waveNew)):
    if i%N == 0:
        dropList = np.append(dropList, i)
waveNew3 = np.delete(waveNew3, dropList)
t2 = np.linspace(0,length/freq,length*freq)
t3 = np.delete(t2, dropList)
len(waveNew3), len(t3)
```

```python
fig = plt.figure(figsize= (20,5))
ax = plt.subplot(111)

#ax.scatter(t1, wave, s=50, c='g', marker='o', linewidths=0, alpha=0.5)
ax.scatter(t2, waveNew+0.2, s=50, c='r', marker='o', linewidths=0, alpha=0.5)
ax.scatter(t2, waveNew2+0.4, s=50, c='y', marker='o', linewidths=0, alpha=0.5)
ax.scatter(t3, waveNew3+0.6, s=50, c='g', marker='o', linewidths=0, alpha=0.5)

ax.set_xlabel("time (sec)")
ax.set_ylabel("y")

ax.set_xlim([1, 1.5])
```
